```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import scatter_matrix

from astropy.table import Table
```

```python
table_RACS_catalogue_URL = 'http://oal.ul.pt/data/AlmaQA2/Catalogs_cV1ax509ZbXXx978pb3/RACS_VEXAS_ALLWISE_DES_PS1_SkyMapper_specZ_milliquas_v3.fits.gz'
table_RACS_cat = Table.read(table_RACS_catalogue_URL)

table_RACS_cat_df = table_RACS_cat.to_pandas() # convert to Pandas Dataframe
```

RACS Survey

```python
attributes=["flux_int","flux_peak","flux_int_err","mean_background","background_noise","n_pix","beam_area"]


scatter_matrix(table_RACS_cat_df[attributes],figsize=(21,14))
```

VEXAS

```python
attributes = ["Y_VISTA","J_VISTA","H_VISTA","KS_VISTA","PSTAR_VISTA","W1MAG","W2MAG","W3MAG","W4MAG"]

scatter_matrix(table_RACS_cat_df[attributes],figsize=(27,18))
```

DES catalog

```python
attributes = ["MAG_AUTO_G_DES","MAG_AUTO_R_DES","MAG_AUTO_I_DES","MAG_AUTO_Z_DES","MAG_AUTO_Y_DES","z_spec_1"]

scatter_matrix(table_RACS_cat_df[attributes],figsize=(27,18))
```

PS Catalog

```python
attributes = ["GPETMAG_PS","RPETMAG_PS","IPETMAG_PS","ZPETMAG_PS","YPETMAG_PS","IPSFMAG_PS","z_spec_2"]

scatter_matrix(table_RACS_cat_df[attributes],figsize=(27,18))
```

SKYMAPPER Catalog

```python
attributes = ["U_PETRO_SM","V_PETRO_SM","G_PETRO_SM","R_PETRO_SM","I_PETRO_SM","Z_PETRO_SM","z_spec_3"]

scatter_matrix(table_RACS_cat_df[attributes],figsize=(27,18))
```
